```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from ipywidgets import interact, FloatSlider
import seaborn as sns

#  seaborn 
sns.set(style="white")

def plot_2d_normal_dist(mu_1=0, mu_2=0, sigma_1=1, sigma_2=1, rho=0):
    """
    
    
    :
    mu_1, mu_2: 
    sigma_1, sigma_2: 
    rho: 
    """
    #  x, y 
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    
    # 
    Z = 0.5/(2 * np.pi * sigma_1 * sigma_2 * np.sqrt(1 - rho**2)) * np.exp(
        -1/(2 * (1 - rho**2)) * (
            (X - mu_1)**2 / sigma_1**2 +
            (Y - mu_2)**2 / sigma_2**2 -
            2 * rho * (X - mu_1) * (Y - mu_2) / (sigma_1 * sigma_2)
        )
    )
    
    # 
    plt.figure(figsize=(8, 6))
    
    plt.contourf(X, Y, Z, 100, cmap='jet', norm=Normalize(vmin=0, vmax=1.3*np.max(Z)))
    plt.colorbar(label='Probability Density')
    plt.xlabel('$X_1$')
    plt.ylabel('$X_2$')
    plt.title('2D Normal Distribution Heatmap')
    plt.show()

#  interact 
interact(plot_2d_normal_dist,
         mu_1=FloatSlider(min=-5, max=5, step=0.1, value=0, description='mu_1'),
         mu_2=FloatSlider(min=-5, max=5, step=0.1, value=0, description='mu_2'),
         sigma_1=FloatSlider(min=0.1, max=5, step=0.1, value=1, description='sigma_1'),
         sigma_2=FloatSlider(min=0.1, max=5, step=0.1, value=1, description='sigma_2'),
         rho=FloatSlider(min=-1, max=1, step=0.01, value=0, description='rho'))
```

```python
import torch
import torch.nn.functional as F
from matplotlib import pyplot as plt
import torch
import numpy as np

# 
import torch.distributions as dist


torch.manual_seed(0)
def gauss(N, A):
    
    #  (a, b)
    a = torch.randint(0, N, (1,)).item()
    b = torch.randint(0, N, (1,)).item()

    #  (sigma_x, sigma_y)
    sigma_x = 8 * torch.rand((1,)).item()
    sigma_y = 8 * torch.rand((1,)).item()

    #  20x20 
    x = torch.arange(0, N, 1, dtype=torch.float32)
    y = x.unsqueeze(1)
    x0 = a
    y0 = b

    # 
    g = A * torch.exp(-((x-x0)**2 / (2 * sigma_x**2) + (y-y0)**2 / (2 * sigma_y**2)))
    return g

N = 40
k = 20

# prob_matrix = 10 * (gauss(N, 1) + gauss(N, 1))  
prob_matrix = torch.tensor(np.load('CANN_U.npy'))
prob_matrix = torch.roll(prob_matrix, shifts=(70, 70), dims=(0, 1))

plt.imshow(prob_matrix, cmap='jet')
plt.colorbar(label='Probability')
plt.xlabel('Column')
plt.ylabel('Row')
plt.title('2D Matrix Heatmap')
plt.show()

# reshape
prob_vector = prob_matrix.view(-1)

# Gumbel-Softmax
prob_vector = prob_vector - prob_vector.max()
softmax_vector = torch.exp(k * prob_vector)
softmax_vector = softmax_vector / softmax_vector.max()

# reshape
softmax_matrix = softmax_vector.view_as(prob_matrix)

plt.imshow(softmax_matrix, cmap='jet')
plt.colorbar(label='Probability')
plt.xlabel('Column')
plt.ylabel('Row')
plt.title('2D Matrix Heatmap')
plt.show()
```

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np
from matplotlib import pyplot as plt
# Parameters for the torus (ring)
R = 0.5  # Major radius of the torus (distance from the center of the tube to the center of the torus)
r = 0.125  # Minor radius of the torus (radius of the tube)
N = 100  # Number of points for generating the torus
delta_x = 0.4
delta_y = 0.7
# Generate torus
theta = np.linspace(0, 2*np.pi, N)
phi = np.linspace(0, 2*np.pi, N)
theta, phi = np.meshgrid(theta, phi)
X = delta_x + (R + r*np.cos(phi)) * np.cos(theta)
Y = delta_y + (R + r*np.cos(phi)) * np.sin(theta)
Z = r * np.sin(phi) + 0.4 * Y  # Tilt the torus similarly to the circle

# Create the figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot torus
# To plot a torus as a collection of polygons, we first need to construct the vertices for each polygon
verts = []

for j in range(len(phi) - 1):
    for i in range(len(theta) - 1):
        verts.append([[X[i, j], Y[i, j], Z[i, j]],
                      [X[i+1, j], Y[i+1, j], Z[i+1, j]],
                      [X[i+1, j+1], Y[i+1, j+1], Z[i+1, j+1]],
                      [X[i, j+1], Y[i, j+1], Z[i, j+1]]])

# Create a Poly3DCollection
NN = len(verts)
colors = []
colors.extend([plt.cm.plasma((0.15 * i) / (NN // 2)) for i in range(0, NN // 2)])
colors.extend([plt.cm.plasma((0.15 * (NN - i)) / (NN // 2)) for i in range(NN // 2, NN)])

# poly = Poly3DCollection(verts, facecolors='red', linewidths=1, edgecolors='b', alpha=.75)
poly = Poly3DCollection(verts, facecolors=colors, edgecolors=None)
# Add collection to the axis
ax.add_collection3d(poly)

# Hide the grid
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Only show the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import seaborn as sns
from ipywidgets import interact, FloatSlider

#  seaborn 
sns.set(style="white")

# 
params1 = {"mu_1": -1, "mu_2": 1, "sigma_1": 2, "sigma_2": 2, "rho": -0.5}
params2 = {"mu_1": 2, "mu_2": 2, "sigma_1": 2, "sigma_2": 2, "rho": 0.5}

def plot_weighted_normal_dist(mu):
    """
    
    
    :
    mu: 
    """
    #  x, y 
    x = np.linspace(-10, 10, 200)
    y = np.linspace(-10, 10, 200)
    X, Y = np.meshgrid(x, y)
    
    # 
    Z1 = 0.5 / (2 * np.pi * params1['sigma_1'] * params1['sigma_2'] * np.sqrt(1 - params1['rho']**2)) * np.exp(
        -1 / (2 * (1 - params1['rho']**2)) * (
            (X - params1['mu_1'])**2 / params1['sigma_1']**2 +
            (Y - params1['mu_2'])**2 / params1['sigma_2']**2 -
            2 * params1['rho'] * (X - params1['mu_1']) * (Y - params1['mu_2']) / (params1['sigma_1'] * params1['sigma_2'])
        )
    )
    
    # 
    Z2 = 0.5 / (2 * np.pi * params2['sigma_1'] * params2['sigma_2'] * np.sqrt(1 - params2['rho']**2)) * np.exp(
        -1 / (2 * (1 - params2['rho']**2)) * (
            (X - params2['mu_1'])**2 / params2['sigma_1']**2 +
            (Y - params2['mu_2'])**2 / params2['sigma_2']**2 -
            2 * params2['rho'] * (X - params2['mu_1']) * (Y - params2['mu_2']) / (params2['sigma_1'] * params2['sigma_2'])
        )
    )
    
    # 
    Z = mu * Z1 + (1 - mu) * Z2
    
    # 
    plt.figure(figsize=(10, 8))
    plt.contourf(X, Y, Z, 100, cmap='viridis', norm=Normalize(vmin=0, vmax=1.3 * np.max(Z)))
    plt.colorbar(label='Probability Density')
    plt.xlabel('$X$')
    plt.ylabel('$Y$')
    plt.title('Weighted 2D Normal Distribution Heatmap')
    plt.show()

#  interact 
interact(plot_weighted_normal_dist,
         mu=FloatSlider(min=0, max=1, step=0.01, value=0.5, description='mu'))
```
